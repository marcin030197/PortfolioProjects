```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(r"C:\Users\marci\Documents\deaths_cases.csv",sep=';')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df[['US_day_cases','US_day_deaths']]
df
```

```python
df['US_day_deaths'] = df['US_day_deaths']
df
```

```python
X = np.array(df.drop(['US_day_deaths'],1))
X
```

```python
y = np.array(df['US_day_deaths'])
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
predictions
```
